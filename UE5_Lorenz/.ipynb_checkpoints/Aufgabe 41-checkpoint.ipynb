```octave
##Aufgabe 41 Punkt c)
function result = mononomialBaseInterpolation(minVal, maxVal)
    result = (1:(maxVal - minVal));
    i = minVal;
    while [i <= maxVal]
    result(i+minVal*(-1)+1) = (103/10) - (19/4)*i + (11/20)*i*i;
    i = i+1;
    endwhile
    #for i = [1:length(result)]
    #    result(i) = (103/10) - (19/4)*i + (11/20)*i*i;
    #sendfor
    return;
endfunction;
result1 = mononomialBaseInterpolation(-25, 25);
plot(result1,'-')
```

```octave
function result = lagrangeInterpolation(minVal, maxVal)
    result = (1:(maxVal - minVal));
    i = minVal;
    while [i <= maxVal]
    result(i+minVal*(-1)+1) = 3*((i-7)*(i-10))/((2-7)*(2-10)) + 2*((i-2)*(i-10))/((7-2)*(7-10)) + 4*((i-2)*(i-7))/((10-2)*(10-7));
    i = i+1;
    endwhile
    #for i = [1:length(result)]
    #    result(i) = 3*((i-7)*(i-10))/((2-7)*(2-10)) + 2*((i-2)*(i-10))/((7-2)*(7-10)) + 4*((i-2)*(i-7))/((10-2)*(10-7));
   # endfor
    return;
endfunction;
result2 = lagrangeInterpolation(-25, 25);
plot(result2, '-')
```
